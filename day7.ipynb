```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
a=pd.read_csv('Melbourne_housing_FULL.csv')
a
```

```python
a.columns
```

```python
a=a.drop(columns=['Address','Date','YearBuilt','Lattitude','Longtitude','Postcode'])
a.head()
```

```python
a.shape
```

```python
a.info()
```

```python
a=a.dropna()
```

```python
a.nunique()
```

```python
a=pd.get_dummies(a,dtype='int')
a.head()
```

```python
X=a.drop(columns=['Price'])
Y=a['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train.values,y_train.values)
```

```python
model.score(x_train.values,y_train.values)
```

```python
model.score(x_test.values,y_test.values)
```

```python
from sklearn.linear_model import Lasso
model2=Lasso(alpha=22,max_iter=100,tol=0.1)
model2.fit(x_train.values,y_train.values)
```

```python
model2.score(x_train.values,y_train.values)
```

```python
model.score(x_test.values,y_test.values)
```
